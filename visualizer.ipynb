```python
import pinocchio as pin
from pinocchio.visualize import MeshcatVisualizer
from pinocchio.robot_wrapper import RobotWrapper
import ipywidgets as widgets
from IPython.display import display
import meshcat
import time
import numpy as np
```

```python
urdf_path = "/home/mmlab/codes/huangshzh/my_mujoco/urdf/dora2_erect_arm.urdf"
mesh_dir = "/home/mmlab/codes/huangshzh/my_mujoco/meshes_7Dof"
robot = RobotWrapper.BuildFromURDF(urdf_path, mesh_dir, pin.JointModelFreeFlyer())
model = robot.model
collision_model = robot.collision_model
visual_model = robot.visual_model

data = robot.data
```

```python
viz = MeshcatVisualizer(model, collision_model, visual_model)
viz.initViewer(loadModel=True, viewer=meshcat.Visualizer())
```

```python
while(1):
    q = pin.randomConfiguration(model)
    q[:6] = [0, 0, 0, 0, 0, 0]
    viz.display(q)
    time.sleep(0.01)
    print(q)
```
